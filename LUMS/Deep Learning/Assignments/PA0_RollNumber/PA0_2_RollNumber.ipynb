# Massachusetts Road Dataset Directory Structure

After downloading and extracting the Massachusetts Road Dataset, the directory structure will look as follows:

1. label_class_dict.csv
2. metadata.csv
3. tiff


Inside the tiff folder you will find the following items:

1. **`test/`**:
   - A set of images used for evaluating the model's performance after training.

2. **`test_labels/`**:
   - Ground-truth labels for the test images, which can be used for validation during the evaluation phase.

3. **`train/`**:
   - The primary set of images used to train the machine learning or deep learning model.

4. **`train_labels/`**:
   - Ground-truth labels for the training images, required for supervised learning tasks.

5. **`val/`**:
   - Images used for validation during the training process to monitor the model's performance and avoid overfitting.

6. **`val_labels/`**:
   - Ground-truth labels for the validation images.

7. **`label_class_dict.csv`**:
   - A CSV file containing the mapping of label classes to their respective identifiers. For example, it might map classes like "road" to numeric values.

9. **`metadata.csv`**:
   - A CSV file providing additional information about the dataset, such as file origins, resolutions, or annotation details.

## Declare Directories


Lets visualize one image from train directory and train_labels directory


```python
# Write Code here
path_to_train = None
path_to_train_labels = None
```

```python
import matplotlib.pyplot as plt
import cv2
import os

image_files = os.listdir(path_to_train)
mask_files = os.listdir(path_to_train_labels)

if image_files and mask_files:
    image_path = os.path.join(path_to_train, image_files[0])  # Get the path of the first image
    mask_path = os.path.join(path_to_train_labels, mask_files[0])      # Get the path of the first mask

    try:
      img = cv2.imread(image_path)
      mask = cv2.imread(mask_path)

      # Display the image and mask
      plt.figure(figsize=(10, 5))

      plt.subplot(1, 2, 1)
      plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB for display
      plt.title('Image')

      plt.subplot(1, 2, 2)
      plt.imshow(cv2.cvtColor(mask, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB for display
      plt.title('Mask')

      plt.show()
    except Exception as e:
      print(f"Error loading or displaying images: {e}")
else:
    print("No images found in the specified directories.")
```
